```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return x ** 2
def g(x):
  return 2 * x
```

```python
# numero aleatorio
x0 = np.random.randint(-5, 5)

# alpha
alpha = 0.5


# lista guardar los saltos sobre la funci√≥n
X = [x0]
Y = [f(x0)]
n_iter = 1000
x = x0
for _ in range(n_iter):

  x = x - alpha * g(x)
  X.append(x)
  Y.append(f(x))
x = np.linspace(-5, 5, 1000)
y = f(x)

plt.plot(x, y)
plt.plot(X, Y, "-o", color="red")
```
